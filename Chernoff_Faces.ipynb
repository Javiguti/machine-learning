```python
from ChernoffFace import *
import numpy
import matplotlib.cm
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#lee el csv, yo lo he hecho desde el raw despues de subirlo a github, pero lo puedes leer desde tu ordenador
#mi csv lo he conseguido al usar esta pagina (https://pulipulichen.github.io/jieba-js/weka/arff2csv/), que te cambia de arff a csv
#he utilizado el original, sin oversampling del minority class, pero si hay que replace missing values que con missing values no funciona lo de las faces
df = pd.read_csv(
    'https://raw.githubusercontent.com/Javiguti/machine-learning/main/practical%20application%203/csv_result-hepatitis_with_no_missing.csv'
)
#dropeo la columna de id (que te genera WEKA) y de Class (al ser non-supervised), y la ultima linea de codigo cambio los valores de male por 1 y female por 0, no hace falta hacerlo
df.drop('id', inplace=True, axis=1)
df.drop('Class', inplace=True, axis=1)
df.SEX=df.SEX.map(dict(male=1, female=0))
```

```python
#para ver que no hay missing values
print(df.head())
```

```python
#si te sale una cosa fea prueba a poner trying= df.head(20), que usaria solo los 20 primeros individuos
#ve cambiando el valor dentro de df.head(), el figsize y el dpi hasta que se muertren todos, o elige mostrar solo 50 por ejemplo como se muestra aqui
trying=df.head(50)
fig = chernoff_face(data=trying,
                    titles=[str(x) for x in list(range(len(trying)))],
                    n_columns=5,
                    long_face=False,
                    color_mapper=matplotlib.cm.tab20b,
                    figsize=(20, 20), dpi=400)
# esta linea la nevesitas para que te salga bonito
fig.tight_layout()
matplotlib.pyplot.show()
```
